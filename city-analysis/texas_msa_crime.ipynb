```python
df_crime.columns = ["City","2007", "2008", "2009", "2010", "2011","2012","2013","2014"]
df_crime.head(3)
```

```python
for row in df_crime:
    df_crime["2015"]=round((df_crime["2014"]-df_crime["2013"]+df_crime["2013"]-df_crime["2012"]+df_crime["2012"]-df_crime["2011"])/3+df_crime["2014"])
#     if (df_crime["2015"]<=0):
#         df_crime["2015"]=df_crime["2014"]
        
    df_crime["2016"]=round((df_crime["2015"]-df_crime["2014"]+df_crime["2014"]-df_crime["2013"]+df_crime["2013"]-df_crime["2012"])/3+df_crime["2015"])
    df_crime["2017"]=round((df_crime["2016"]-df_crime["2015"]+df_crime["2015"]-df_crime["2014"]+df_crime["2014"]-df_crime["2013"])/3+df_crime["2016"])

df_crime["Location"]= df_crime.City.str.split(" Pol").str.get(0)
# df_crime
```

```python
df_ratings["Location"]= df_ratings.name.str.split(",").str.get(0)
df_ratings.head(3)
```

```python
df_crimed=pd.merge(df_ratings, df_crime, how="left", on="Location")
df_crime=df_crimed.drop(["Location", "City"], axis=1) 
df_crime.columns = ["City","Rating","2007", "2008", "2009", "2010", "2011","2012","2013","2014","2015","2016","2017"]
df_crime
```

```python
df_crime["join"]=df_crime.index
df_crime
```

```python
df_crime["join"][3]=2
df_crime["join"][11]=10
df_crime["join"][12]=10
df_crime["join"][15]=14
df_crime["join"][16]=14
df_crime["join"][29]=28
df_crime["join"][23]=22
df_crime["join"][24]=22
df_crime["join"][18]=17
df_crime["join"][7]=6
df_crime["join"][5]=4
df_crime["join"][31]=30

df_crime_com=df_crime.groupby(["join"]).sum()
```

```python
df_crime_com["City"]="X"
df_crime_com["City"][2]="Austin-Round Rock, TX"
df_crime_com["City"][10]="Dallas-Fort Worth-Arlington, TX"
df_crime_com["City"][14]="Houston-The Woodlands-Sugar Land, TX"
df_crime_com["City"][28]="San Antonio-New Braunfels, TX"
df_crime_com["City"][22]="McAllen-Edinburg-Mission, TX"
df_crime_com["City"][17]="Killeen-Temple, TX"
df_crime_com["City"][6]="Brownsville-Harlingen, TX"
df_crime_com["City"][4]="Beaumont-Port Arthur, TX"
df_crime_com["City"][9]="Corpus Christi, TX"
df_crime_com["City"][13]="El Paso, TX"

df_crime_com
```

```python
df_crime_com=df_crime_com.drop(["Rating"], axis=1)
```

```python
df_crime_com=df_crime_com[df_crime_com.City != 'X']
# df_crime_com=df_crime_com.set_index('City')
df_crime_com
```

```python
# df_crime_2=df_crime_com.pivot_table(index=["City","2007", "2008", "2009", "2010", "2011", "2012", "2013", "2014" ,"2015" ,"2016" ,"2017" ])

city_crime=pd.melt(df_crime_com, 
            id_vars='City', 
            value_vars=["2007", "2008", "2009", "2010", "2011", "2012", "2013", "2014" ,"2015" ,"2016" ,"2017"] # list of days of the week
            )
city_crime
```

```python
city_crimer=city_crime.sort_values(by=['City', 'variable'])
city_crimer.columns = ["City","Year", "Crime"]
city_crimer
```

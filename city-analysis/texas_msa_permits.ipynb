```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
csvpath = "original_data/dataPermit_full_annual.csv"
msa_df = pd.read_csv(csvpath)
msa_df.shape
```

```python
# cleared out all other states and cities other than Texas
msa_texas_df = msa_df[msa_df['area'].str.contains(", TX")]
msa_texas_df.shape
```

```python
# cleared out all years other than 2007-2017
msa_texas_year_df = msa_texas_df[msa_texas_df.year >= 2007]
msa_texas_year_df.shape
```

```python
# adding up the three columns of "units" per permit and creating a new total permits column
msa_texas_year_df["total permits"] = msa_texas_year_df.f1units + msa_texas_year_df.f24units + msa_texas_year_df.f5units
msa_texas_year_df
```

```python
# dropping all columns we do not need
msa_texas_total = msa_texas_year_df.drop(
    [
    "f1units", 
    "f1change", 
    "f1value", 
    "f1valchange", 
    "f24units", 
    "f24change", 
    "f24value", 
    "f24valchange", 
    "f5units", 
    "f5change", 
    "f5value", 
    "f5valchange"
    ]
    , axis = 1
)
```

```python
# setting the index to Area with sub of year
msa_texas_total.set_index(['area', 'year'], inplace=True)
```

```python
msa_texas_total.head()
```

```python
msa_texas_total.to_csv("output_data/texas_permits.csv", index=True, header=True)
```

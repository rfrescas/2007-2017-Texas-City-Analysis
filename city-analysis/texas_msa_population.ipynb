```python
# Dependencies
import csv
import matplotlib.pyplot as plt
import requests
import pandas as pd
# from config import api_key
```

```python
# File to Load (Remember to Change These)
file_to_read = "original_data/dataPop_full.csv"

# Read Purchasing File and store into Pandas data frame
pop_df = pd.read_csv(file_to_read)
pop_df.tail()
```

```python
pop_df.shape
```

```python
pop_tx_df = pop_df[pop_df["area"].str.contains(", TX")]
pop_tx_df.shape
```

```python
pop_tx_df.tail()
```

```python
pop_tx_years_df = pop_tx_df[pop_tx_df.year >= "2007"]
pop_tx_years_df.shape
```

```python
pop_tx_years_df.head()
```

```python
pop_tx_years_df.dtypes
```

```python
pop_tx_years_df["population"] = pop_tx_years_df["population"].astype(int)
```

```python
pop_tx_years_df.dtypes
```

```python
pop_tx_years_df.head()
```

```python
pop_tx_table = pd.pivot_table(pop_tx_years_df, values="population", index=['area','year'])
pop_tx_table
```

```python
area_unique = pop_tx_years_df.groupby("area").mean()
area_unique
```

```python
pop_tx_table.to_csv("output_data/texas_population.csv", index=True, header=True)
```

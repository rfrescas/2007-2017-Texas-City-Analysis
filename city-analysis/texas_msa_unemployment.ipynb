```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# csv file saved in same folder
csv_unemployment = "original_data/dataEmpl_full_annual.csv"

# create df from csv, clean and display first rows
texas_unemployment = pd.read_csv(csv_unemployment)
texas_unemployment.drop(["employmentTotal","employmentChange","unemploymentChange"],axis=1,inplace=True)
texas_unemployment = texas_unemployment[texas_unemployment.year > 2006]
texas_unemployment = texas_unemployment[texas_unemployment.year < 2018]
texas_unemployment = texas_unemployment.rename(columns={"unemploymentRate":"unemployment rate"})
texas_unemployment = texas_unemployment[texas_unemployment.area.str.contains(", TX")]
texas_unemployment.head()
```

```python
#create a multi-index df for visualization purposes only
unemployment_table = pd.pivot_table(texas_unemployment, values='unemployment rate', index=['area','year'])
unemployment_table.head()
```

```python
# another way to display df for visualization puposes only
employment_table2 = pd.pivot_table(texas_unemployment, values='unemployment rate',index="area",columns='year').reset_index()
employment_table2.head()
```

```python
unemployment_table.to_csv("output_data/texas_unemployment.csv", index=True, header=True)
```
